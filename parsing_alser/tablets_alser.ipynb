```python
import requests
from bs4 import BeautifulSoup
import csv


HOST = 'https://www.alser.kz'
URL  = 'https://alser.kz/c/vse-planshety'
HEADERS = {'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.193 Safari/537.36'
          }


def get_html(url):                                
    r = requests.get(url)
    return r

def write_csv(data):
    with open('tablets_alser_kz.csv', 'a', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow((
                          data['title'],
                          data['price'],
                          data['brand'],
                          data['model'],
                          data['list'],
                          data['size'],
                          data['processor'],
                          data['frequency'],
                          data['volume'],
                          data['main_camera'],
                          data['front_camera'],
                          data['ram']
                           
                         ))

                         

def get_content(html):
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_='product-item')


    for tablet in items:
        url_info = HOST+tablet.find('a', class_='product-item__info_title').get('href')
        html_info = get_html(str(url_info))
        tablet_info = BeautifulSoup(html_info.text, 'html.parser')
        item = tablet_info.find('div', class_='detail-specifications').find_all('p')
        print(url_info)
        try:
            title = tablet.find('a', class_='product-item__info_title').get_text(strip=True)
        except:
            title = None
        try:
            price = tablet.find('div', class_='price').get_text(strip=True)
        except:
            price = None
           
        try:
            brand = item[1].get_text(strip=True)
        except:
            brand = None
        try:
            model = item[5].get_text(strip=True)
        except:
            model = None
        try:
            list =  item[3].get_text(strip=True)
        except:
            list = None
        
        try:
            size = item[13].get_text(strip=True)
        except:
            size = None
        try:
            processor = item[23].get_text(strip=True)
        except:
            processor = None
        try:
            frequency = item[29].get_text(strip=True)
        except:
            frequency = None
        try:
            volume = item[33].get_text(strip=True)
        except:
            volume = None
        try:
            main_camera = item[57].get_text(strip=True)
        except:
            main_camera = None
        try:
            front_camera=item[59].get_text(strip=True)
        except:
            front_camera=None
        try:
            ram = item[31].get_text(strip=True)
        except:
            ram = None
        
            

        tablets = {
            'title': title,
            'price': price,
            'brand': brand,
            'model': model,
            'list': list,
            'size': size,
            'processor': processor,
            'frequency': frequency,
            'volume': volume,
            'main_camera': main_camera,
            'front_camera': front_camera,
            'ram': ram
            
            
        }
        write_csv(tablets)
        

if __name__=='__main__':
    page_part = '?page='

    total_pages = 4

    for i in range(1, total_pages):
        print(i)
        url_gen = URL + page_part + str(i)
        print(url_gen)
        html = get_html(url_gen)
        get_content(html.text)
```
