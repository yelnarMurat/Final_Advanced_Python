```python
# importing necessary libiries 
import requests
from bs4 import BeautifulSoup
import csv

 

HOST = 'https://www.alser.kz' #main website
URL  = 'https://alser.kz/c/vse-televizory'  # main webpage
HEADERS = {'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.193 Safari/537.36'
          }


def get_html(url):                 #function for getting html page
    r = requests.get(url)
    return r

def write_csv(data):
    with open('televizor_alser_kz.csv', 'a', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow((
                          data['title'],
                          data['price'],
                          data['brand'],
                          data['model'],
                          data['list'],
                          data['size'],
                          data['resolution'],
                          data['smart_tv'],
                          data['wifi']
                           
                        
        ))

                         

def get_content(html):                                 #function for getting content of the webpage
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_='product-item')


    for tv in items:
        url_info = HOST+tv.find('a', class_='product-item__info_title').get('href')
        html_info = get_html(str(url_info))
        tv_info = BeautifulSoup(html_info.text, 'html.parser')
        item = tv_info.find('div', class_='detail-specifications').find_all('p')
        print(url_info)
        try:
            title = tv.find('a', class_='product-item__info_title').get_text(strip=True)
        except:
            title = None
        try:
            price = tv.find('div', class_='price').get_text(strip=True)
        except:
            price = None
           
        try:
            brand = item[1].get_text(strip=True)
        except:
            brand = None
        try:
            model = item[5].get_text(strip=True)
        except:
            model = None
        try:
            list =  item[3].get_text(strip=True)
        except:
            list = None
        
        try:
            size = item[9].get_text(strip=True)
        except:
            size = None
        try:
            resolution = item[11].get_text(strip=True)
        except:
            resolution = None
        try:
            smart_tv = item[25].get_text(strip=True)
        except:
            smart_tv = None
        try:
            wifi = item[45].get_text(strip=True)
        except:
            wifi = None
        
            

        tvs = {
            'title': title,
            'price': price,
            'brand': brand,
            'model': model,
            'list': list,
            'size': size,
            'resolution': resolution,
            'smart_tv': smart_tv,
            'wifi': wifi
            
            
        }
        write_csv(tvs)
        

if __name__=='__main__':                           #indicating number of pages
    page_part = '?page='

    total_pages = 6

    for i in range(1, total_pages):
        print(i)
        url_gen = URL + page_part + str(i)
        print(url_gen)
        html = get_html(url_gen)
        get_content(html.text)
```
