```python
# importing necessary libiries 
import requests
from bs4 import BeautifulSoup
import csv



HOST = 'https://www.alser.kz'   #main website
URL  = 'https://alser.kz/c/vse-noutbuki'   # main webpage
HEADERS = {'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'
          }


def get_html(url):          #function for getting html page
    r = requests.get(url)
    return r

def write_csv(data):
    with open('laptopss_alser_kz.csv', 'a', encoding='utf8') as f:
        writer = csv.writer(f)
        writer.writerow(( data['title'],
                          data['price'],
                          data['brand'],
                          data['list'],
                          data['size'],
                          data['os'],
                          data['model_cpu'],
                          data['freq'],
                          data['ram_size'],
                          data['hard_typee'],
                          data['storage'],
                          data['videocard'],
                          data['videocard_mem']
                         ))

def get_content(html):                                      #function for getting content of the webpage
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_='product-item')


    for laptop in items:
        url_info = HOST+laptop.find('a', class_='product-item__info_title').get('href')
        html_info = get_html(str(url_info))
        laptop_info = BeautifulSoup(html_info.text, 'html.parser')
        item = laptop_info.find('div', class_='detail-specifications').find_all('p')
        print(url_info)
        try:
            title = laptop.find('a', class_='product-item__info_title').get_text(strip=True)
        except:
            title = None
        try:
            price = laptop.find('div', class_='price').get_text(strip=True)
        except:
            price = None
           
        try:
            brand = item[35].get_text(strip=True)
        except:
            brand = None
        
        try:
            list =  item[37].get_text(strip=True)
        except:
            list = None
        
        try:
            size = item[51].get_text(strip=True)
        except:
            size = None
        try:
            os = item[25].get_text(strip=True)
        except:
            os = None
        try:
            model_cpu = item[5].get_text(strip=True)
        except:
            model_cpu = None
        try:
            freq = item[9].get_text(strip=True)
        except:
            freq = None
        try:
            ram_size = item[23].get_text(strip=True)
        except:
            ram_size = None
        try:
            hard_typee = item[45].get_text(strip=True)
        except:
            hard_typee = None
        try:
            storage = item[47].get_text(strip=True)
        except:
            storage = None
        try:
            videocard = item[17].get_text(strip=True)
        except:
            videocard = None
        try:
            videocard_mem = item[19].get_text(strip=True)
        except:
            videocard_mem = None
            

        laptops = {
            'title': title,
            'price': price,
            'brand': brand,
            'list': list,
            'size': size,
            'os': os,
            'model_cpu': model_cpu,
            'freq': freq,
            'ram_size': ram_size,
            'hard_typee': hard_typee,
            'storage': storage,
            'videocard': videocard,
            'videocard_mem': videocard_mem
        }
        write_csv(laptops)
    
    
if __name__=='__main__':                     #indicating number of pages
    page_part = '?page='

    total_pages = 7

    for i in range(1, total_pages):
        print(i)
        url_gen = URL + page_part + str(i)
        print(url_gen)
        html = get_html(url_gen)
        get_content(html.text)
```
