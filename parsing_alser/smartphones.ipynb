```python
# importing necessary libiries 
import requests
from bs4 import BeautifulSoup
import csv



HOST = 'https://www.alser.kz'  #main website
URL  = 'https://alser.kz/c/vse-smartfony'  # main webpage
HEADERS = {'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
          }


def get_html(url):                #function for getting html page
    r = requests.get(url)
    return r

def write_csv(data):
    with open('smartphones_alser_kzz.csv', 'a', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(( data['title'],
                          data['price'],
                          data['brand'],
                          data['model'],
                          data['list'],
                          data['size'],
                          data['os'],
                          data['ram_size'],
                          data['volume'],
                          data['main_camera'],
                          data['front_camera'],
                          data['videshoot_format'],
                          data['videoshoot_resolution']
                
                           ))

def get_content(html):                           #function for getting content of the webpage
    soup = BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_='product-item')


    for smartphone in items:
        url_info = HOST+smartphone.find('a', class_='product-item__info_title').get('href')
        html_info = get_html(str(url_info))
        smartphone_info = BeautifulSoup(html_info.text, 'html.parser')
        item = smartphone_info.find('div', class_='detail-specifications').find_all('p')
        print(url_info)
        try:
            title = smartphone.find('a', class_='product-item__info_title').get_text(strip=True)
        except:
            title = None
        try:
            price = smartphone.find('div', class_='price').get_text(strip=True)
        except:
            price = None
           
        try:
            brand = item[1].get_text(strip=True)
        except:
            brand = None
        try:
            model = item[5].get_text(strip=True)
        except:
            model = None
        try:
            list =  item[3].get_text(strip=True)
        except:
            list = None
        
        try:
            size = item[21].get_text(strip=True)
        except:
            size = None
        try:
            os = item[27].get_text(strip=True)
        except:
            os = None
        try:
            ram_size = item[39].get_text(strip=True)
        except:
            ram_size = None
        try:
            volume = item[47].get_text(strip=True)
        except:
            volume = None
        try:
            main_camera = item[51].get_text(strip=True)
        except:
            main_camera = None
        try:
            front_camera = item[53].get_text(strip=True)
        except:
            front_camera = None
        try:
            videoshoot_format = item[61].get_text(strip=True)
        except:
            videoshoot_format = None
        try:
            videoshoot_resolution = item[63].get_text(strip=True)
        except:
            videoshoot_resolution = None
            

        smartphones = {
            'title': title,
            'price': price,
            'brand': brand,
            'model': model,
            'list': list,
            'size': size,
            'os': os,
            'ram_size': ram_size,
            'volume': volume,
            'main_camera': main_camera,
            'front_camera': front_camera,
            'videshoot_format': videoshoot_format,
            'videoshoot_resolution': videoshoot_resolution
            
        }
        write_csv(smartphones)
        

if __name__=='__main__':                   #indicating number of pages
    page_part = '?page='

    total_pages = 15

    for i in range(1, total_pages):
        print(i)
        url_gen = URL + page_part + str(i)
        print(url_gen)
        html = get_html(url_gen)
        get_content(html.text)
```
